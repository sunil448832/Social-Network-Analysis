```
from networkx import barabasi_albert_graph
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from itertools import combinations
from random import random
```

```
def Random_graph(n,e):
  V=[v for v in range(n)]
  E= [i for i in combinations(V,2)]
  np.random.shuffle(E)
  E=E[:e]
  G = nx.Graph()
  G.add_nodes_from(V)
  G.add_edges_from(E)
  return G

def Random_graph_p(n, p):
    vertex=[v for v in range(n)]
    V=set(vertex)
    E=set()
    for comb in combinations(V, 2):
        x = random()
        if x < p:E.add(comb)
    G = nx.Graph()
    G.add_nodes_from(V)
    G.add_edges_from(E)
    return G

print("random graph generation with #nodes n and #egdes e")
G1 = Random_graph(100,300)
nx.draw(G1,with_labels=False)
plt.draw()
```

```
print("random graph generation with #nodes n and egdes probablity p")
G2 = Random_graph_p(100,.4)
nx.draw(G2,with_labels=False)
plt.draw()
```

```
print("random graph generation with barabasi_albert model")
G=barabasi_albert_graph(600,400)
nx.draw(G,with_labels=False)
plt.draw()
```

```

degrees = G.degree()
n_nodes={}
for i in degrees:
  try:n_nodes[i[1]]+=1
  except:n_nodes[i[1]]=1

degree=[]
nodes=[]
log_nodes=[]
log_degree=[]
for k ,v in sorted(n_nodes.items(), key=lambda x: x[0]):
  degree.append(k)
  nodes.append(v)
  log_degree.append(np.log10(k))
  log_nodes.append(np.log10(v))

plt.plot(degree,nodes,label='Degree distribution')
plt.legend()
plt.xlabel('Degree of node x')
plt.ylabel('No of node with degree=x')
plt.show()
plt.plot(log_degree,log_nodes,label='lograthmic Degree distribution')
plt.legend()
plt.xlabel('Degree of node x')
plt.ylabel('No of node with degree=x')
plt.show()
```
